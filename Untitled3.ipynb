```python
# Analiza Salariilor in Regim Remote - Google Colab Notebook

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Incarcare date
df = pd.read_csv("/content/Work_From_Anywhere_Salary_Data.csv")

# Previzualizare
print(df.head())
print(df.info())

# Statistici descriptive
print(df.describe())

# Verificare valori lipsa
print(df.isnull().sum())

# Distributia salariilor
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary (Annual)'], kde=True, bins=30, color='skyblue')
plt.title('Distributia Salariilor Anuale')
plt.xlabel('Salariu (Anual)')
plt.ylabel('Frecventa')
plt.show()

# Boxplot - salariu in functie de nivelul de experienta
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Experience Level', y='Salary (Annual)', palette='Set2')
plt.title('Salariu dupa Nivelul de Experienta')
plt.show()

# Salariu mediu pe industrie
industry_salary = df.groupby('Industry')['Salary (Annual)'].mean().sort_values()
plt.figure(figsize=(10, 8))
industry_salary.plot(kind='barh', color='teal')
plt.title('Salariu Mediu pe Industrie')
plt.xlabel('Salariu Mediu (Anual)')
plt.ylabel('Industrie')
plt.show()

# Comparatie: Remote vs Onsite
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Remote Flexibility', y='Salary (Annual)', palette='coolwarm')
plt.title('Salariu in functie de Flexibilitatea Remote')
plt.show()

# Satisfactia medie pe Tech Stack
tech_stack_satisfaction = df.groupby('Tech Stack')['Job Satisfaction Score (1-10)'].mean().sort_values()
plt.figure(figsize=(10, 8))
teach_stack_plot = tech_stack_satisfaction.plot(kind='barh', color='orange')
plt.title('Satisfactia Medie pe Tech Stack')
plt.xlabel('Scor Mediu Satisfactie')
plt.ylabel('Tech Stack')
plt.show()

# Tipuri de angajare
employment_type_counts = df['Employment Type'].value_counts()
plt.figure(figsize=(6, 6))
employment_type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distributia Tipurilor de Angajare')
plt.ylabel('')
plt.show()

# Corelatii intre variabile numerice
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matricea de Corelatie')
plt.show()
```
